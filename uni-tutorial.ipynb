```python
#Generate discrete time points(X values)
X = np.linspace(0,0.1,500)

#Create subplots(2X2 grid)
fig, ax = plt.subplots(2,2, figsize=(10,8))

#Plot 5sin(100piX) in the first subplot
ax[0,0].plot(X,5*np.sin(100*np.pi*X), label = r'$5\sin(100\pi X)$', color = 'b')
ax[0,0].grid(True)

#Plot 5sin(100piX+pi/8) in the second subplot
ax[0,1].plot(X,5*np.sin(100*np.pi*X+np.pi/8), label = r'$5\sin(100\pi X + \frac{\pi}{8}$', color = 'g')
ax[0,1].grid(True)

#Plot cos(100piX) in the third subplot
ax[1,0].plot(X, np.cos(100*np.pi*X), label = r'$\cos(100\pi X)$', color = 'r')
ax[1,0].grid(True)

#Plot cos(100piX-pi) in the fourth subplot
ax[1,1].plot(X, np.cos(100*np.pi*X-np.pi), label = r'$\cos(100\pi X - \pi)$', color = 'm')
ax[1,1].grid(True)

for i in range(len(ax)):
    for j in range(len(ax[0])):
        ax[i,j].legend()
        ax[i,j].set_xlabel('Time (s)')
        ax[i,j].set_ylabel('Amplitude')

fig.suptitle('Different Wave Functions')
plt.tight_layout()
plt.show()        
```

```python
def draw(frequency, amplitude,duration, sampling_rate):
    X = np.linspace(0,duration,sampling_rate*duration)
    plt.plot(X, amplitude*np.sin(2*np.pi*frequency*X), label = fr"${amplitude}\sin({2*frequency}\pi X)$", c = 'r')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid(True)
    plt.title('Sine Wave Signal')
    plt.show()
```

```python
draw(10,2,1,1000)
```

```python
def draw2(frequency, amplitude, duration, sampling_rate, discrete_amplitude):
    X = np.linspace(0,duration, sampling_rate*duration)

    if (discrete_amplitude == False):
        plt.stem(X, amplitude*np.sin(2*np.pi*frequency*X), label=fr"${amplitude}\sin({2*frequency}\pi X)$")
    else:
        plt.stem(X, (amplitude*np.sin(2*np.pi*frequency*X)).astype(int), label = fr"${amplitude}\sin({2*frequency}\pi X)$")

    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid(True)
    plt.title('Discrete Sine Wave Signal')
    plt.show()        
```

```python
draw2(duration=1, amplitude=3, frequency=5, discrete_amplitude=True, sampling_rate=100)
```

```python
draw2(duration=1, amplitude=3, frequency=5, discrete_amplitude=False, sampling_rate=100)
```

```python
def discrete(duration, sampling_rate, transition_probability):
    a = np.random.uniform(0,1,sampling_rate*duration)
    for i in range(len(a)):
        if a[i] >= 1-transition_probability:
            a[i] = 1
        else:
            a[i] = 0
    X = np.linspace(0,duration,sampling_rate*duration)
    y = 1
    Y = np.zeros(duration*sampling_rate)
    for i in range(len(Y)):
        if a[i] == 0:
           Y[i] = y
        else:
            Y[i] = -1*y
            y = -1*y

    plt.step(X,Y)
    plt.title('Digital Signal') 
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.show()          
```

```python
discrete(duration=2,sampling_rate=15,transition_probability=0.1)
```

```python
def discrete_digital(duration, sampling_rate, transition_probability):
    #Generate random values to decide where transition will occur
    transitions = np.random.uniform(0,1, int(sampling_rate*duration))

    #Initialize the amplitude array of the signal, starting with 1
    signal = np.ones(int(sampling_rate*duration))

    #Toggle the signal amplitude based on the transition probability between -1 and 1 
    for i in range(len(transitions)):
        if transitions[i] < transition_probability:
            signal[i] = -1*signal[i-1] # TOGGLE AMPLITUDE
        else:
            signal[i] = signal[i-1]

    time = np.linspace(0,duration,int(sampling_rate*duration))

    #plot the digital signal
    plt.step(time,signal, where='mid')
    plt.title('Digital-Like Signal(Amplitude: ±1)')    
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show()    

discrete_digital(duration=2, sampling_rate=15, transition_probability=0.2)
```
