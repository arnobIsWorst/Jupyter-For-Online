```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
X_data = np.random.random(100) * 100
Y_data = np.random.random(100) * 100

plt.scatter(X_data, Y_data, s = 150, alpha=0.3)
plt.show()
```

```python
years = [2006 + x for x in range(19)]
np.random.seed(1)
weights = np.random.uniform(70, 100, size=19)

plt.plot(years, weights, lw = 3)
plt.xticks(years[::2])
plt.show()
```

```python
x = ["C++", "C", "Java", "Python", "Rust"]
y = [20, 50, 130, 40, 45]

plt.bar(x, y, color = "skyblue")
plt.show()
```

```python
ages = np.random.normal(20, 7.5, 1000)
plt.hist(ages, bins = 20, cumulative=True)
plt.show()
```

```python
langs = ["C++", "C#", "Java", "Python", "Rust"]
votes = [20, 50, 100, 4, 85]
explode = (0, 0, 0.1, 0.1, 0.1)

plt.pie(votes, labels = langs, autopct = "%.2f%%", pctdistance= 0.7, explode=explode)
plt.show()
```

```python
heights = np.random.normal(172, 8, 300)
plt.boxplot(heights)
plt.show()
```

```python
first = np.linspace(0, 10, 25)
second = np.linspace(10, 200, 25)
third = np.linspace(200, 210, 25)
fourth = np.linspace(210, 230, 25)

data = np.concatenate((first, second, third, fourth))
plt.boxplot(data)
plt.show()
```

```python
years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
income = [110, 275, 339, 77, 557, 420, 701, 889, 900, 850, 1099]
income_ticks = list(range(50, 1200, 100))

plt.plot(years, income)
plt.title("Income of John over the years(in USD)")
plt.xlabel("Years")
plt.ylabel("Yearly Income in USD")
plt.yticks(income_ticks, [f"${x}k" for x in income_ticks])

plt.show()

```

```python
stock_a = [100, 102, 99, 101, 101, 100, 102]
stock_b = [98, 95, 102, 104, 105, 103, 109]
stock_c = [110, 115, 100, 105, 100, 98, 95]

plt.plot(stock_a, label = "Company A")
plt.plot(stock_b, label = "Company B")
plt.plot(stock_c, label = "Company C")
plt.legend()   

plt.show()
```

```python
votes = [10, 2, 5, 16, 22]
people = ["John", "Alice", "Bob", "Diana", "George"]

style.use("dark_background") 
plt.pie(votes, labels=None, autopct = "%.1f%%")
plt.legend(people,)

plt.show()
```

```python
x1, y1 = np.random.random(100), np.random.random(100)
x2, y2 = np.arange(100), np.random.random(100)

style.use("default")
plt.figure(1)
plt.scatter(x1, y1)

plt.figure(2)
plt.plot(x2, y2)

plt.show()
```

```python
x = np.arange(100)

fig, axs = plt.subplots(2, 2) #2 rows 2 columns

axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("Sine wave")

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title("Cosine wave")

axs[1, 0].plot(x, np.random.random(100))
axs[1, 0].set_title("Random function")

axs[1, 1].plot(x, np.log(x))
axs[1, 1].set_title("Logarithmic function")

fig.suptitle("Different functions")
plt.tight_layout()
plt.show()
```

```python
ax = plt.axes(projection = "3d")  

x = np.arange(0, 50, 0.1)
y = np.sin(x)
z = np.cos(x)

ax.scatter(x, y, z)
ax.set_title("3D Plot")

plt.show()
```

```python
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)

X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

ax.plot_surface(X, Y, Z, cmap = "Spectral")
ax.set_title("3D Surface Plot")

plt.show()
```

```python
import random

heads_tails = [0, 0]

for _ in range(1000):
    heads_tails[random.randint(0, 1)] += 1
    plt.bar(["Heads", "Tails"], heads_tails, color = ["skyblue", "orange"])
    plt.pause(0.001)

plt.show()    
```
